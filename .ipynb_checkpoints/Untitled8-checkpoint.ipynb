```python
def wc(filename):
    with open(filename,"r") as file:
        lines=file.readlines()

    line_count=len(lines)
    word_count=0
    for i in lines:
        word_count+=len(i.split())
    char_count=sum(len(line) for line in lines)
    return line_count, word_count, char_count
print(wc("sample.txt"))
        
```

```python
def safe_divide(a,b):
    try:
        return a/b
    except (ZeroDivisionError, TypeError):
        return None
print(safe_divide(10,"a"))
```

```python
def read_lines(file_name):
    try:
        with open(file_name, "r") as file:
            reader=file.readlines()
            return reader
    except FileNotFoundError:
        print(f"'{file_name}' not found")
        return []
print(read_lines("notes.txt"))
print(read_lines("missing.txt"))
```

```python
import json

# Create a valid JSON file
with open("valid.json", "w") as file:
    json.dump({"name": "Alice", "age": 30}, file)

# Create a broken JSON file (just text that isn't valid JSON)
with open("bad.json", "w") as file:
    file.write("{not valid json")
```

```python
def safe_parse_json(filename):
    try:
        with open(filename, "r") as file:
            return json.load(file)
    except FileNotFoundError:
        print(f"File '{filename}' not found")
        return None
    except json.JSONDecodeError:
        print("Invalid JSON.")
        return None
print(safe_parse_json("valid.json"))       # {'name': 'Alice', 'age': 30}
print(safe_parse_json("bad.json"))         # prints "Invalid JSON.", returns None
print(safe_parse_json("missing.json"))     # prints "File not found.", returns None
```

```python
def validate_age(age):
    if not isinstance(age, int):
        raise ValueError("Age must be an integer")
    elif age<0 or age>120:
        raise ValueError("Age must be between 0 and 120")
    else:
        return age
print(validate_age(30))

try:
    validate_age(-5)
except ValueError as e:
    print(f"Caught: {e}")

try:
    validate_age(200)
except ValueError as e:
    print(f"Caught: {e}")

try:
    validate_age("thirty")
except ValueError as e:
    print(f"Caught: {e}")
```

```python
class Counter:
    def __init__(self, start=0):
        self.value=start
    def increment(self):
        self.value+=1
    def reset(self):
        self.value=0
    def decrement(self):
        self.value-=1
c=Counter()
c.increment()
c.increment()
c.increment()
print(c.value)
c.reset()
print(c.value)
c2=Counter(start=10)
c2.increment()
c2.increment()
c2.decrement()
print(c2.value)
print(c.value)
```

```python
class Box:
    def __init__(self, size):
        
        self.size = size
        self.contents = []
    
    def add(self, item):
        self.contents.append(item)
    
    def is_full(self):
        return len(self.contents) >= self.size

box1 = Box(3)
box2 = Box(5)

box1.add("apple")
box1.add("banana")

box2.add("hammer")

print(box1.contents)       # ?
print(box2.contents)       # ?
print(box1.is_full())      # ?
print(box2.is_full())      # ?

box1.add("cherry")
print(box1.is_full())      # ?
```

```python
class BankAccount:
    def __init__(self, owner, balance=100):
        self.owner=owner
        self.balance=balance
    def deposit(self, amount):
        self.balance+=amount
    def withdraw(self, amount):
        if self.balance>amount:
            self.balance-=amount
        else:
            raise ValueError("insufficient funds")
```

```python
account1=BankAccount("shashank", 10000000)
account2=BankAccount("sara", 10000000)
account1.deposit(20000)
account2.deposit(10000)
account2.withdraw(5000)
print(account1.balance)
print(account2.balance)

try:
    account1.withdraw(200000000)
except ValueError as e:
    print(f"caught {e}")
```

```python
import csv

class Library:
    def __init__(self, filename="library.csv"):
        self.filename = filename
        self.books = []
        self._load()
    
    def _load(self):
        try:
            with open(self.filename, "r") as file:
                reader = csv.DictReader(file)
                for row in reader:
                    row["year"] = int(row["year"])
                    row["read"] = row["read"] == "True"
                    self.books.append(row)
        except FileNotFoundError:
            pass
    def _save(self):
        with open(self.filename, "w", newline="") as file:
            writer = csv.DictWriter(file, fieldnames=["title", "author", "genre", "year", "read"])
            writer.writeheader()
            writer.writerows(self.books) 
        

    
    def add_book(self):
        print("\n-----add book------")
        title = input("title: ").strip()
        author = input("author: ").strip()
        genre = input("genre: ").strip()
        
        while True:
            try:
                year = int(input("year: "))
                break
            except ValueError:
                print("Not a valid number")
        
        read_input = input("Have you read it? (y/n): ").strip().lower()
        read = read_input in ("y", "yes")
        
        book = {
            "title": title,
            "author": author,
            "genre": genre,
            "year": year,
            "read": read
        }
        self.books.append(book)
        self._save()
        print(f"Added '{title}' to library")

    def _display_book(self, book):
        status = "✓" if book["read"] else " "
        print(f"[{status}] {book['title']} by {book['author']} ({book['year']}) — {book['genre']}")

        
    def view_books(self):
        print("\n-----your list-----")
        if not self.books:
            print("library is empty")
            return
        for book in self.books:
            self._display_book(book)
    
    def find_book(self, title):
        title = title.strip().lower()
        for index, book in enumerate(self.books):
            if book["title"].lower() == title:
                return index
        return None             
    
    def search_books(self):
        print("\n----search-----")
        print("Search by 1=title, 2=author, 3=genre, 4=year")
        field_choice=input("choice: ")
    
        field_map={"1":"title","2":"author","3":"genre","4":"year"}
        if field_choice not in field_map:
            print("invalid choice")
            return
        field=field_map[field_choice]
        term=input(f"Search {field} : ").lower()
    
        matches=[book for book in self.books if term in book[field].lower()]
        print(f"\n found {len(matches)} matches: ")
        for book in matches:
            self._display_book(book)
    
    def mark_as_read(self):
        print("\n--- Mark as Read ---")
        title = input("Title of book to mark as read: ")
        index = self.find_book(title)
        
        if index is None:
            print(f"No book titled '{title}' found.")
            return
        
        self.books[index]["read"] = True
        self._save()
        print(f"Marked '{self.books[index]['title']}' as read.")
    
    def delete_book(self):
        print("\n--- Delete a Book ---")
        title = input("Title of book to delete: ")
        index = self.find_book(title)
        
        if index is None:
            print(f"No book titled '{title}' found.")
            return
        
        removed = self.books.pop(index)
        self._save()
        print(f"Deleted '{removed['title']}'.")
    
    def show_stats(self):
        print("\n--- Library Stats ---")
    
        if not self.books:
            print("(library is empty)")
            return
        total = len(self.books)
        read_count = sum(1 for book in self.books if book["read"])
        unread_count = total - read_count
        
        print(f"Total books: {total}")
        print(f"Read: {read_count}")
        print(f"Unread: {unread_count}")


        
def main():
    library = Library()
    while True:
        print("\n=== Book Library Manager ===")
        print("1. Add a book")
        print("2. View all books")
        print("3. Search books")
        print("4. Mark as read")
        print("5. Delete a book")
        print("6. View stats")
        print("7. Quit")
        
        choice = input("Choice: ")
        
        if choice == "1":
            library.add_book()
        elif choice == "2":
            library.view_books()
        elif choice == "3":
            library.search_books()
        elif choice == "4":
            library.mark_as_read()
        elif choice == "5":
            library.delete_book()
        elif choice == "6":
            library.show_stats()
        elif choice == "7":
            print("Goodbye!")
            break
        else:
            print(f"Invalid choice: {choice}")

main()
```
