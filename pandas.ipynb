```python
## Data validation note

The Chipotle dataset's `item_price` column is the **line total** (already multiplied by quantity), 
NOT the unit price. Many tutorials and Day 2 of this sprint incorrectly compute revenue as 
`item_price * quantity`, which double-counts multi-quantity rows.

Correct: `revenue = item_price` (use as-is)
Wrong: `revenue = item_price * quantity`

Impact: ~10% over-count across the dataset. Top 5 items by revenue stay the same, but absolute 
numbers drop. Discovered Day 3 during pandas migration.
```

```python
# Step 1: re-read and clean (in case state is messy)
df = pd.read_csv("C:/Users/shash/chipotle.tsv", sep="\t")
df["item_price"] = df["item_price"].str.replace("$", "").str.strip().astype(float)

# Step 2: CORRECT revenue formula — item_price is already the line total
df["revenue"] = df["item_price"]

# Step 3: top 5 by total revenue, correctly
top_5 = (df.groupby("item_name")
           .agg(total_revenue=("revenue", "sum"))
           .sort_values("total_revenue", ascending=False)
           .head(5))
print(top_5)
```

```python
total_revenue = df["revenue"].sum()
unique_orders = df["order_id"].nunique()
avg_per_order = round(total_revenue / unique_orders, 2)
print(f"Correct avg revenue per order: ${avg_per_order}")
```
