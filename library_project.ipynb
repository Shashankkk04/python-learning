```python
import csv
def load_library():
    library=[]
    try:
        with open("library.csv","r") as file:
            reader=csv.DictReader(file)
            for row in reader:
                row["year"]=int(row["year"])
                row["read"]=row["read"]=="True"
                library.append(row)
    except FileNotFoundError:
        pass
    return library



def save_library(library):
    with open("library.csv", "w", newline="") as file:
        writer=csv.DictWriter(file, fieldnames=["title","author","genre","year","read"])
        writer.writeheader()
        writer.writerows(library)

def safe_int_input(prompt):
    while True:
        num=input(prompt)
        try:
            return int(num)
        except ValueError:
            print("Not a valid number")
            continue


def add_book(library):
    print("\n-----add book------")
    title = input("Title: ").strip()
    author = input("Author: ").strip()
    genre = input("Genre: ").strip()
    year = safe_int_input("Year: ")
    read_input = input("Have you read it? (y/n): ").strip().lower()
    read = read_input in ("y", "yes")

    book={
        "title":title,
        "author": author,
        "genre": genre,
        "year": year,
        "read": read
    }
    library.append(book)
    save_library(library)
    print(f"Added '{title}' to library")



def display_book(book):
    status = "✓" if book["read"] else " "
    print(f"[{status}] {book['title']} by {book['author']} ({book['year']}) — {book['genre']}")



def view_books(library):
    print("\n-----your list-----")
    if not library:
        print("library is empty")
        return
    for book in library:
        display_book(book)



def search_books(library):
    print("\n----search-----")
    print("Search by 1=title, 2=author, 3=genre, 4=year")
    field_choice=input("choice: ")

    field_map={"1":"title","2":"author","3":"genre","4":"year"}
    if field_choice not in field_map:
        print("invalid choice")
        return
    field=field_map[field_choice]
    term=input(f"Search {field} : ")

    matches=[book for book in library if term in book[field].lower()]
    print(f"\n found {len(matches)} matches: ")
    for book in matches:
        display_book(book)


def find_book(library, title):
    """Return the index of the first book matching title (case-insensitive). None if not found."""
    title = title.strip().lower()
    for index, book in enumerate(library):
        if book["title"].lower() == title:
            return index
    return None
    

def mark_as_read(library):
    print("\n--- Mark as Read ---")
    title = input("Title of book to mark as read: ")
    index = find_book(library, title)
    
    if index is None:
        print(f"No book titled '{title}' found.")
        return
    
    library[index]["read"] = True
    save_library(library)
    print(f"Marked '{library[index]['title']}' as read.")




def delete_book(library):
    print("\n--- Delete a Book ---")
    title = input("Title of book to delete: ")
    index = find_book(library, title)
    
    if index is None:
        print(f"No book titled '{title}' found.")
        return
    
    removed = library.pop(index)
    save_library(library)
    print(f"Deleted '{removed['title']}'.")

def show_stats(library):
    print("\n--- Library Stats ---")
    
    if not library:
        print("(library is empty)")
        return
    
    total = len(library)
    read_count = sum(1 for book in library if book["read"])
    unread_count = total - read_count
    
    print(f"Total books: {total}")
    print(f"Read: {read_count}")
    print(f"Unread: {unread_count}")
    
    # Count by genre
    by_genre = {}
    for book in library:
        genre = book["genre"]
        by_genre[genre] = by_genre.get(genre, 0) + 1
    
    print("\nBy genre:")
    for genre, count in by_genre.items():
        print(f"  {genre}: {count}")
    
    # Unique authors
    authors = {book["author"] for book in library}
    print(f"\nUnique authors: {len(authors)}")
    print("  " + ", ".join(authors))
    
    # Books read by year
    by_year = {}
    for book in library:
        if book["read"]:
            year = book["year"]
            by_year[year] = by_year.get(year, 0) + 1
    
    print("\nBooks read by year:")
    if not by_year:
        print("  (none read yet)")
    else:
        for year, count in sorted(by_year.items()):
            print(f"  {year}: {count}")

def main():
    library=load_library()
    while True:
        print("\n=== Book Library Manager ===")
        print("1. Add a book")
        print("2. View all books")
        print("3. Search books")
        print("4. Mark as read")
        print("5. Delete a book")
        print("6. View stats")
        print("7. Quit")

        choice=input("choice: ")
        if choice=="1":
            add_book(library)
        elif choice=="2":
            view_books(library)
        elif choice=="3":
            search_books(library)
        elif choice == "4":
            mark_as_read(library)
        elif choice == "5":
            delete_book(library)
        elif choice == "6":
            show_stats(library)
        elif choice == "7":
            print("good bye !")
            break
        else:
            print(f"you picked option {choice} - not built yet")

main()
```
